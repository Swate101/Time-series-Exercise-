 ## 6. Resample by the day and calculate the min and max temp for the day (Hint: .agg(['min', 'max'])).

```python
min_max_temp = T.resample('D').agg(['min', 'max'])
min_max_temp     # this shows the min_max as you can see in the below display 
```

 ## 6a. Use this resampled dataframe to calculate the change in temperature for the day.

```python
min_max_temp['temp_diff'] = T.resample('D').max() - T.resample('D').min()# As you can see the changes in min and max
min_max_temp.info()
min_max_temp.head()
```

```python
min_max_temp['temp_diff'] = T.resample('D').max() - T.resample('D').min() # As you can see the changes in min and max
min_max_temp.info()
```

## 6b. Which month has the highest daily temperature variability?
- This answer  is still September one of the hottest months 

```python
min_max_temp.resample('M').temp_diff.mean().idxmax()
```

## Bonus: Visualize the daily min, average, and max temperature over time on a single line plot, i.e. the min, average, and maximum temperature should be 3 seperate lines.

```python
T.resample('D').temp.min().plot(label = 'min')
T.resample('D').temp.mean().plot(label = 'mean')
T.resample('D').temp.max().plot(label = 'max')
plt.legend()
```

```python
T.resample('D').agg(['min','max','mean']).plot()
plt.title('Temperature Over Time by Days')
plt.ylabel('Temperature')
plt.legend(('Min','Max','Avg'))          # I like this way better it just feels cleaner 
```
